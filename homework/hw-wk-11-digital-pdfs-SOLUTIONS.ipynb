```python
## import unzipper

import glob
import tabula
import pandas as pd
import itertools
```

```python
## gather all pdfs in a list
myfiles = sorted(glob.glob('all_homework*/*FUDS*.pdf'))
myfiles
```

```python
## iterate to pull out tables using tabula
df_tables = []
for myfile in myfiles:
    df_list = tabula.read_pdf(myfile, pages = "all")
    df_tables.append(df_list)
    
```

#### The next few cells just show the nest quality of all the lists generated

```python
## 3 items in this list. each item is a list that holds the tables from each pdf
len(df_tables) 
```

```python
## remember that tabula returns a list with pandas df in them for each table in the pdf
## so we have a list for each pdf that holds all the dfs for each pdf
len(df_tables[0])
```

```python
## nested list
type(df_tables[0])
```

```python
## the main list holds lists, with the nested lists holding dataframes
type(df_tables[0][0])
```

## Use itertools to flatten the lists

```python
## flatten list of lists to a single list with all the dataframes
all_tables = list(itertools.chain(*df_tables))
len(all_tables)
```

```python
## now each item is a df
type(all_tables[0])
```

```python
## concat all the dataframes into one, reset index
df = pd.concat(all_tables).reset_index(drop = True)
df
```
